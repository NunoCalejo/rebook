```python
import os
import binarize
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Current Pre-processing - Best Binarization

### from: https://github.com/phulin/rebook/blob/master/rebook/binarize.py 

```python
directory = r'.\Camera'

width = 500
height = 500
page_number = 0
for filename in os.listdir(directory):
    
    img = cv2.imread(r'.\Camera\\' + filename ,0)
    resized = cv2.resize(img, (width, height), interpolation = cv2.INTER_NEAREST)
    sav = binarize.sauvola(resized) 
    
    cv2.imwrite(r'.\Results\\multiple' + str(page_number) + '.png', sav)
    
    print('Page Count:' + str(page_number+1) , end='\r')
    page_number += 1 
```
